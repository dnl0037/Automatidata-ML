### **Task 2c. Understand the data - Investigate the variables**



Sorting `trip_distance` from maximum to minimum value:

```python
df.sort_values(by="trip_distance",ascending=False)[["passenger_count","trip_distance","fare_amount","tip_amount","total_amount"]]
```

* The values align with our earlier data discovery, where we noticed that the longest rides are approximately 33 miles.


Sorting by `total_amount`:

```python
df.sort_values(by="total_amount",ascending=False).head(5)
```

* The first two values are significantly higher than the others.

Creating `trip_duration` column and sorting it by `total_amount`:

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

df['trip_duration'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']
```

```python
df[["trip_duration","total_amount"]].sort_values(by="total_amount",ascending=False).head()
```

Sorting the data by `total_amount`:

```python
df[["trip_duration","total_amount"]].sort_values(by="total_amount").head()
```

Getting the representation of each payment type:

```python
df[['payment_type']].value_counts()
```

According to the data dictionary, the payment method was encoded as follows:

1 = Credit card  
2 = Cash  
3 = No charge  
4 = Dispute  
5 = Unknown  
6 = Voided trip

Obtaining the average tip for trips paid for with credit card and cash:

```python
print(f'Average tip for credit card: {df[df["payment_type"]==1]["tip_amount"].mean():.2f}')
print(f'Average tip for cash: {df[df["payment_type"]==2]["tip_amount"].mean()}')
```

Times is each vendor ID represented in the data:

```python
df["VendorID"].value_counts()
```

Getting the mean `total_amount` for each vendor:

```python
df.groupby(["VendorID"])[["total_amount"]].mean()
```

From the data points payed with credit card, getting the trip count for every `passenger_count:

```python
credit_card = df[df["payment_type"]==1]
credit_card[["passenger_count"]].value_counts().reset_index().rename(columns={0:"# trips"})
```

Calculating the average tip amount for each passenger count (credit card payments only):

```python
credit_card.groupby(["passenger_count"])[["tip_amount"]].mean().rename(columns={"tip_amount":"tip_amount_mean"})
```
